```python
#plot a polygon.  Input vertices in cyclic order
def plotpoly(p, figsize=(6,6), ax=None):
    # stolen from JOSH #geometry
    if ax == None:
        fig, ax = plt.subplots(1,1, figsize=figsize)
    for i in range(len(p)-1):
        ax.plot([p[i][0],p[i+1][0]],[p[i][1],p[i+1][1]], marker='o')
    ax.plot([p[len(p)-1][0],p[0][0]],[p[len(p)-1][1],p[0][1]])

def plot_dyck(path, figsize=(6,6), ax=None):
    '''plot a single dyck path'''
    x = 0
    y = 0
    points = [[x,y]]
    for i in path:
        if i == 1:
            y += 1
            x += 1
        else:
            y -= 1
            x += 1
        points.append([x,y])
    plotpoly(points, figsize, ax)
    
def plot_paths(dyckset, cols=4, Stat=None):
    '''plot many dyck paths'''
    rows = len(dyckset) // cols + 1
    f, axs = plt.subplots(rows, cols)
    f.set_figheight(rows * 3)
    f.set_figwidth(cols * 3)
    for i in range(len(dyckset)):
        # print(i//cols, i%cols)
        ax = axs[i//cols][i%cols]
        plot_dyck(dyckset[i], ax=ax)
        ax.axis('off')
        ax.set_ylim([0,rows+0.5])
        if Stat != None:
            ax.set_title(str(Stat(dyckset[i])), y=-0.15)
    for i in range(rows*cols-len(dyckset)+1):
        axs[-1][-i].axis("off")
```

```python
plot_paths(set1, cols=4, Stat=Stat)
```
